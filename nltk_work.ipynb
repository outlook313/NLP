```python
#Do you know AI?
syn=wordnet.synsets('AI')
syn
```

```python
syn[1].definition()
```

```python
#To get the list of synonyms:
synonyms=[]
for syn in wordnet.synsets('AI'):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
synonyms
```
